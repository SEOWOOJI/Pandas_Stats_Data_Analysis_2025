#_**y = a*x + b**_

### where
### a & b are _constantsÏÉÅÏàò_ (or _coefficientÍ≥ÑÏàò_). The constant '**a**' is referred as a '<font color = 'blue'> **slope** </font>.' The consonat '**b**' is referred as an '<font color = 'blue'> **intercept** </font>.'  



# üéà **About the constant '<font color = 'red'>a</font>'**
# **y = b + <font color = 'red'> a</font>*x**

- ### Varying the value of a slope (aka. a): positive values for Rising üìà vs. negative values for Falling üìâ



# üéà **About the constant '<font color = 'brown'>b</font>'**
# **y =  <font color = 'brown'> b</font> + a*x**
>
- ### If b > 0, the graph intersects with the y-axis above the origin.
- ### If b < 0, the graph intersects with the y-axis below the origin.
- ### If b = 0, the graph intersects with the y-axis at the origin.

    > <font color = 'brown'> The term "origin" refers to the point (0,0) on the Cartesian coordinate plane. "origin"Ïù¥ÎûÄ Ïπ¥ÌÖåÏãúÏïà Ï¢åÌëú ÌèâÎ©¥ ÏÉÅÏùò (0,0) ÏßÄÏ†ê

## 1Ô∏è‚É£ If a > 0 and b > 0; a = 3 and b 2

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-10, 10, 100)

a = 3
b = 2

y = 2 + a*x

plt.plot(x, y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Graph with a > 0 and b > 0')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the range of x values
x = np.linspace(-10, 10, 100) #'x' will be an arry containing 100 values ranging from -10 to 10

# Define parameters
a = 3
b = 2

# Calculate y values
y = b + a * x

# Plot the graph
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Graph with a > 0 and b > 0')
plt.grid(True)
plt.show()
```
